```python
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam

# ─── Configuration ─────────────────────────────────────────────────────────
base_path = "DS" 
subjects = range(1, 16)
runs     = range(1, 11)
eeg_labels = [
    "F3","F1","Fz","F2","F4","FFC5h","FFC3h","FFC1h","FFC2h","FFC4h","FFC6h",
    "FC5","FC3","FC1","FCz","FC2","FC4","FC6","FTT7h","FCC5h","FCC3h","FCC6h",
    "FTT8h","C5","C3","C1","Cz","C2","C4","C6","TTP7h",
    "CCP5h","CCP3h","CCP1h","CCP2h","CCP4h","CCP6h","TTP8h","CP5","CP3","CP1",
    "CPz","CPP5h","CPP3h","CPP1h","CPP2h","CPP4h","CPP6h","P3","P1","Pz","P2","P4","PPO1h",
    "PPO2h"
]
event_codes = list(range(1,8)) 

# ─── 1) Load & epoch all runs ───────────────────────────────────────────────
X_list, y_list = [], []
for subj in subjects:
    for run in runs:
        fp = f"{base_path}/S{subj}_MI/motorimagination_subject{subj}_run{run}.gdf"
        raw = mne.io.read_raw_gdf(fp, preload=True, verbose=False)
        # pick only valid EEG channels
        to_keep = [ch for ch in eeg_labels if ch in raw.ch_names]
        raw.pick_channels(to_keep)
        raw.filter(0.5, 50.0, verbose=False)
        # events and epoch
        events, _ = mne.events_from_annotations(raw)
        epochs = mne.Epochs(
            raw, events, event_codes,
            tmin=-1.0, tmax=1.5,
            baseline=(None, 0.0),
            preload=True, verbose=False
        )
        data = epochs.get_data()          # (n_epochs, n_ch, n_times)
        labels = epochs.events[:,2].astype(int)  # array of 1–7
        X_list.append(data)
        y_list.append(labels)

X = np.concatenate(X_list, axis=0)
y = np.concatenate(y_list, axis=0)
print("Loaded data:", X.shape, "labels:", np.unique(y), np.bincount(y))

# ─── 2) Impute all‐NaN or all‐zero channels by median time series ───────────
n_epochs, n_ch, n_t = X.shape
median_ts = np.nanmedian(X, axis=0)  # (n_ch, n_t)

for i in range(n_epochs):
    for c in range(n_ch):
        ep = X[i, c, :]
        if np.all(np.isnan(ep)) or np.all(ep == 0):
            X[i, c, :] = median_ts[c]

# rolling‐mean fill remaining holes
for i in range(n_epochs):
    for c in range(n_ch):
        ts = X[i, c, :].astype(float)
        ts[(ts == 0) | np.isnan(ts)] = np.nan
        s = pd.Series(ts)
        filled = s.rolling(5, min_periods=1, center=True).mean()
        filled.fillna(filled.median(), inplace=True)
        X[i, c, :] = filled.values

# verify no empty channels
assert not np.any(np.all(np.isnan(X), axis=2))
assert not np.any(np.all(X == 0,      axis=2))

# ─── 3) Min–Max scaling X and y ────────────────────────────────────────────
X_flat = X.reshape(n_epochs, n_ch * n_t)
scalerX = MinMaxScaler()
Xs_flat = scalerX.fit_transform(X_flat)
Xs = Xs_flat.reshape(n_epochs, n_ch, n_t)

y = y.reshape(-1,1)
scalerY = MinMaxScaler()
ys = scalerY.fit_transform(y).ravel()
y_cls = np.round(ys * 6).astype(int)  # now 0–6

# ─── 4) Sliding‐window segmentation ────────────────────────────────────────
def sliding_windows(X, y, window=10, step=5):
    Xw, yw = [], []
    for xi, yi in zip(X, y):
        for start in range(0, xi.shape[1] - window + 1, step):
            Xw.append(xi[:, start:start+window].T)  # transpose to (window, n_ch)
            yw.append(yi)
    return np.array(Xw, dtype='float32'), np.array(yw, dtype='int32')

Xw, yw = sliding_windows(Xs, y_cls, window=10, step=5)
print("Windowed:", Xw.shape, yw.shape)

# ─── 5) Train/Val/Test split ────────────────────────────────────────────────
X_tr, X_tmp, y_tr, y_tmp = train_test_split(
    Xw, yw, test_size=0.4, random_state=42, stratify=yw
)
X_val, X_te, y_val, y_te = train_test_split(
    X_tmp, y_tmp, test_size=0.5, random_state=42, stratify=y_tmp
)
print("Split:", X_tr.shape, X_val.shape, X_te.shape)

# ─── 6) Build & compile two‐layer LSTM ─────────────────────────────────────
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_tr.shape[1], X_tr.shape[2])),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(7, activation='softmax')
])
model.compile(optimizer=Adam(1e-3),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# ─── 7) Train & validate ──────────────────────────────────────────────────
history = model.fit(
    X_tr, y_tr,
    validation_data=(X_val, y_val),
    epochs=100, batch_size=32, verbose=2
)

# ─── 8) Final evaluate & plot ─────────────────────────────────────────────
test_loss, test_acc = model.evaluate(X_te, y_te, batch_size=32, verbose=0)
print(f"\nTest loss: {test_loss:.4f} | Test accuracy: {test_acc:.2%}")

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],   label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'],   label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy'); plt.legend()
plt.tight_layout()
plt.show()
```

```python
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.optimizers import Adam

# ─── Configuration ─────────────────────────────────────────────────────────
base_path = "DS"  # adjust to your dataset root
subjects = range(1, 16)
runs     = range(1, 11)
eeg_labels = [
    "F3","F1","Fz","F2","F4","FFC5h","FFC3h","FFC1h","FFC2h","FFC4h","FFC6h",
    "FC5","FC3","FC1","FCz","FC2","FC4","FC6","FTT7h","FCC5h","FCC3h","FCC6h",
    "FTT8h","C5","C3","C1","Cz","C2","C4","C6","TTP7h",
    "CCP5h","CCP3h","CCP1h","CCP2h","CCP4h","CCP6h","TTP8h","CP5","CP3","CP1",
    "CPz","CPP5h","CPP3h","CPP1h","CPP2h","CPP4h","CPP6h","P3","P1","Pz","P2","P4","PPO1h",
    "PPO2h"
]
event_codes = list(range(1,8))  # numeric codes 1–7 in GDF

# ─── 1) Load & epoch all runs ───────────────────────────────────────────────
X_list, y_list = [], []
for subj in subjects:
    for run in runs:
        fp = f"{base_path}/S{subj}_MI/motorimagination_subject{subj}_run{run}.gdf"
        raw = mne.io.read_raw_gdf(fp, preload=True, verbose=False)
        # pick only valid EEG channels
        to_keep = [ch for ch in eeg_labels if ch in raw.ch_names]
        raw.pick_channels(to_keep)
        raw.filter(0.5, 50.0, verbose=False)
        # events and epoch
        events, _ = mne.events_from_annotations(raw)
        epochs = mne.Epochs(
            raw, events, event_codes,
            tmin=-1.0, tmax=1.5,
            baseline=(None, 0.0),
            preload=True, verbose=False
        )
        data = epochs.get_data()          # (n_epochs, n_ch, n_times)
        labels = epochs.events[:,2].astype(int)  # array of 1–7
        X_list.append(data)
        y_list.append(labels)

X = np.concatenate(X_list, axis=0)
y = np.concatenate(y_list, axis=0)
print("Loaded data:", X.shape, "labels:", np.unique(y), np.bincount(y))

# ─── 2) Impute all‐NaN or all‐zero channels by median time series ───────────
n_epochs, n_ch, n_t = X.shape
median_ts = np.nanmedian(X, axis=0)  # (n_ch, n_t)

for i in range(n_epochs):
    for c in range(n_ch):
        ep = X[i, c, :]
        if np.all(np.isnan(ep)) or np.all(ep == 0):
            X[i, c, :] = median_ts[c]

# rolling‐mean fill remaining holes
for i in range(n_epochs):
    for c in range(n_ch):
        ts = X[i, c, :].astype(float)
        ts[(ts == 0) | np.isnan(ts)] = np.nan
        s = pd.Series(ts)
        filled = s.rolling(5, min_periods=1, center=True).mean()
        filled.fillna(filled.median(), inplace=True)
        X[i, c, :] = filled.values

# verify no empty channels
assert not np.any(np.all(np.isnan(X), axis=2))
assert not np.any(np.all(X == 0,      axis=2))

# ─── 3) Min–Max scaling X and y ────────────────────────────────────────────
X_flat = X.reshape(n_epochs, n_ch * n_t)
scalerX = MinMaxScaler()
Xs_flat = scalerX.fit_transform(X_flat)
Xs = Xs_flat.reshape(n_epochs, n_ch, n_t)

y = y.reshape(-1,1)
scalerY = MinMaxScaler()
ys = scalerY.fit_transform(y).ravel()
y_cls = np.round(ys * 6).astype(int)  # now 0–6

# ─── 4) Sliding‐window segmentation ────────────────────────────────────────
def sliding_windows(X, y, window=10, step=5):
    Xw, yw = [], []
    for xi, yi in zip(X, y):
        for start in range(0, xi.shape[1] - window + 1, step):
            Xw.append(xi[:, start:start+window].T)  # transpose to (window, n_ch)
            yw.append(yi)
    return np.array(Xw, dtype='float32'), np.array(yw, dtype='int32')

Xw, yw = sliding_windows(Xs, y_cls, window=10, step=5)
print("Windowed:", Xw.shape, yw.shape)

# ─── 5) Train/Val/Test split ────────────────────────────────────────────────
X_tr, X_tmp, y_tr, y_tmp = train_test_split(
    Xw, yw, test_size=0.4, random_state=42, stratify=yw
)
X_val, X_te, y_val, y_te = train_test_split(
    X_tmp, y_tmp, test_size=0.5, random_state=42, stratify=y_tmp
)
print("Split:", X_tr.shape, X_val.shape, X_te.shape)

window_size, n_channels = X_tr.shape[1], X_tr.shape[2]
n_classes = len(np.unique(y_tr))

# ─── 1) Build & train a 1D‐CNN ───────────────────────────────────────────────
model_cnn = Sequential([
    Conv1D(32, kernel_size=3, activation='relu',
           input_shape=(window_size, n_channels)),
    MaxPooling1D(pool_size=2),
    Conv1D(64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(n_classes, activation='softmax')
])
model_cnn.compile(
    optimizer=Adam(1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_cnn = model_cnn.fit(
    X_tr, y_tr,
    validation_data=(X_val, y_val),
    epochs=20, batch_size=32, verbose=2
)

# Evaluate CNN
loss_cnn, acc_cnn = model_cnn.evaluate(X_te, y_te, verbose=0)

print(f"CNN Test Loss: {loss_cnn:.4f} | CNN Test Acc: {acc_cnn:.2%}")

# Plot CNN learning curves
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history_cnn.history['loss'],   label='Train Loss')
plt.plot(history_cnn.history['val_loss'], label='Val Loss')
plt.title('CNN Loss'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history_cnn.history['accuracy'],   label='Train Acc')
plt.plot(history_cnn.history['val_accuracy'], label='Val Acc')
plt.title('CNN Accuracy'); plt.legend()
plt.tight_layout()
plt.show()

# ─── 2) Train & evaluate Random Forest ──────────────────────────────────────
# Flatten windows for RF
X_tr_rf = X_tr.reshape(X_tr.shape[0], -1)
X_val_rf= X_val.reshape(X_val.shape[0], -1)
X_te_rf = X_te.reshape(X_te.shape[0], -1)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_tr_rf, y_tr)

y_val_pred = rf.predict(X_val_rf)
y_test_pred = rf.predict(X_te_rf)

val_acc_rf = accuracy_score(y_val, y_val_pred)
test_acc_rf = accuracy_score(y_te, y_test_pred)

print(f"RF Val Acc: {val_acc_rf:.2%} | RF Test Acc: {test_acc_rf:.2%}")

# ─── 3) Summarize comparison ───────────────────────────────────────────────
print("\nModel Comparison:")
print(f"LSTM Test Acc: {test_acc:.2%}") 
print(f"CNN  Test Acc: {acc_cnn:.2%}")
print(f"RF   Test Acc: {test_acc_rf:.2%}")
```
