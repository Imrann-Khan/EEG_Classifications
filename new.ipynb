```python
# List to store all data (EEG data and labels)
all_X = []
all_y = []
```

```python
import numpy as np
import mne

# Assuming you have the list `all_X` and `all_y` initialized

expected_channels = 61  # Set the expected number of EEG channels (in your case, 61)

for subject in range(1, 16):  # 15 subjects
    for run in range(1, 11):  # 10 runs per subject
        file_path = f"{base_path}/S{subject}_MI/motorimagination_subject{subject}_run{run}.gdf"
        print(f"Processing: {file_path}")
        
        # Load EEG data using MNE
        raw = mne.io.read_raw_gdf(file_path, preload=True)
        
        # Keep only the first 61 EEG channels
        # Select the first 61 channels if there are more than 61
        if len(raw.info['ch_names']) > expected_channels:
            # Pick the desired channels
            raw.pick_channels(raw.info['ch_names'][:expected_channels])
            
        # Add check for less than expected channels and raise an exception
        elif len(raw.info['ch_names']) < expected_channels:
            raise ValueError(f"Subject {subject} run {run} has fewer than {expected_channels} channels. Please check your input data.")
            
        # Apply bandpass filter (0.5 to 50 Hz for motor imagery)
        raw.filter(0.5, 50)

        # Define events and event codes (adjust as per your data)
        events, _ = mne.events_from_annotations(raw)

        # Extract epochs (EEG data for each trial)
        epoch = mne.Epochs(raw, events, event_id=[1, 2, 3, 4, 5, 6, 7], tmin=-1.5, tmax=1.0)
        
        # Append EEG data and event labels
        X = epoch.get_data()  # Shape: (n_epochs, n_channels, n_times)
        y = epoch.events[:, 2]  # Event codes as labels
        
        # Append to the overall list
        all_X.append(X)
        all_y.append(y)
```

```python
all_X = np.concatenate(all_X, axis=0)
all_y = np.concatenate(all_y, axis=0)
```

```python
# Scale the data to [0, 1] range
X = all_X.copy()
n_epochs, n_ch, n_t = X.shape

for ch in range(n_ch):
    ch_data = X[:, ch, :]               # shape (n_epochs, n_times)
    ch_mean = np.nanmean(ch_data)       # mean over all epochs & times
    ch_data[np.isnan(ch_data)] = ch_mean
    X[:, ch, :] = ch_data

all_X = X
print("Any NaNs left after imputation?", np.isnan(all_X).any())
```

```python
# mask of epochs without any NaNs
good_epochs = ~np.isnan(all_X).any(axis=(1,2))

# filter both X and y
all_X = all_X[good_epochs]
all_y = all_y[good_epochs]

print(f"Dropped {np.sum(~good_epochs)} epochs with NaNs; remaining:", all_X.shape[0])
print("Any NaNs in all_X after scaling? ", np.isnan(all_X).any())
print("Shape of all_X:", all_X.shape)  # (n_epochs, n_channels, n_times)
```

LSTM Model

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split

#  1) Hyperparameters 
window_size = 10
num_epochs  = 10
num_units   = 50
batch_size  = 32

#  2) Assuming you already have all_X (n_epochs, n_times, n_channels)
#         and all_y (n_epochs,) prepared and scaled

# Split into train / val / test
X_tmp, X_test, y_tmp, y_test = train_test_split(
    all_X, y_all_adjusted, test_size=0.2, random_state=42, stratify=y_all_adjusted
)
X_train, X_val, y_train, y_val = train_test_split(
    X_tmp, y_tmp, test_size=0.25, random_state=42, stratify=y_tmp
)

#  3) Sliding-window generator 
def sliding_windows(X, y, window, step=None):
    if step is None:
        step = window // 2
    Xw, yw = [], []
    for xi, yi in zip(X, y):
        T = xi.shape[0]
        for start in range(0, T - window + 1, step):
            Xw.append(xi[start:start+window])
            yw.append(yi)
    return np.array(Xw, dtype='float32'), np.array(yw, dtype='int32')

# Create windowed sets
X_train_w, y_train_w = sliding_windows(X_train, y_train, window_size)
X_val_w,   y_val_w   = sliding_windows(X_val,   y_val,   window_size)
X_test_w,  y_test_w  = sliding_windows(X_test,  y_test,  window_size)

#  4) Build a two-layer LSTM + Dense model 
n_channels = X_train_w.shape[2]
n_classes  = len(np.unique(y_all_adjusted))+1

model = Sequential([
    # first LSTM returns full sequence
    LSTM(num_units, return_sequences=True,
         input_shape=(window_size, n_channels)),
    Dropout(0.2),
    # second LSTM outputs only final state
    LSTM(num_units, return_sequences=False),
    Dropout(0.2),
    # final Dense for classification
    Dense(n_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

#  5) Train 
history = model.fit(
    X_train_w, y_train_w,
    validation_data=(X_val_w, y_val_w),
    epochs=num_epochs,
    batch_size=batch_size
)

#  6) Evaluate 
test_loss, test_acc = model.evaluate(X_test_w, y_test_w, batch_size=batch_size)
print(f"Test Loss: {test_loss:.4f}  |  Test Acc: {test_acc:.2%}")
```

GA-LSTM Model

```python
import numpy as np
import random
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam

#  0) Data: assumes you already have these arrays ready 
# X_train, X_val, X_test: shape (n_epochs, n_times, n_channels)
# y_train, y_val, y_test: shape (n_epochs,)
num_classes = len(np.unique(y_train))

#  1) GA PARAMS 
POP_SIZE       = 4
NUM_GENERATIONS= 4
WINDOW_BITS    = 6
UNIT_BITS      = 8
EPOCH_BITS     = 6
GENE_LENGTH    = WINDOW_BITS + UNIT_BITS + EPOCH_BITS

#  2) HELPERS 
def decode_gene(gene):
    """Convert bitlist  (window_size, num_units, num_epochs)."""
    # bits 05  window
    w = int("".join(map(str, gene[0:WINDOW_BITS])), 2)
    # bits 613  units
    u = int("".join(map(str, gene[WINDOW_BITS:WINDOW_BITS+UNIT_BITS])), 2)
    # bits 1419  epochs
    e = int("".join(map(str, gene[WINDOW_BITS+UNIT_BITS:])), 2)
    return w, u, e


def sliding_windows(X, y, window, step=None):
    """Chop each trial (n_timesn_ch) into overlapping windows."""
    if step is None:
        step = window // 2
    Xw, yw = [], []
    for xi, yi in zip(X, y):
        T = xi.shape[0]
        for start in range(0, T-window+1, step):
            Xw.append(xi[start:start+window])
            yw.append(yi)
    return np.array(Xw, dtype='float32'), np.array(yw, dtype='int32')

def build_model(window_size, n_channels, num_units, num_classes):
    """Two-layer LSTM  softmax."""
    m = Sequential([
        LSTM(num_units, return_sequences=True,
             input_shape=(window_size, n_channels)),
        Dropout(0.2),
        LSTM(num_units),
        Dropout(0.2),
        Dense(num_classes+1, activation='softmax')
    ])
    m.compile(
        optimizer=Adam(learning_rate=1e-3, clipvalue=1.0),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return m

def fitness(window_size, num_units, num_epochs):
    """Train on X_train/X_val for given hyperparams; return val_loss."""
    # must skip invalid combos
    if window_size == 0 or num_units == 0 or num_epochs == 0:
        return np.inf

    # 1) build windowed datasets
    Xtr, ytr = sliding_windows(X_train, y_train, window_size)
    Xvl, yvl = sliding_windows(X_val,   y_val,   window_size)
    if len(Xtr) == 0 or len(Xvl) == 0:
        return np.inf

    # 2) train & return final val_loss
    m = build_model(window_size, X_train.shape[2], num_units, num_classes)
    h = m.fit(Xtr, ytr,
              validation_data=(Xvl, yvl),
              epochs=num_epochs,
              batch_size=32,
              verbose=0)
    return h.history['val_loss'][-1]

#  3) INITIAL POPULATION 
population = [
    [random.randint(0,1) for _ in range(GENE_LENGTH)]
    for _ in range(POP_SIZE)
]
best_gene = population[0].copy()
best_loss = float('inf')

#  4) GA MAIN LOOP 
for gen in range(NUM_GENERATIONS):
    print(f"\n--- Generation {gen+1}/{NUM_GENERATIONS} ---")
    fitnesses = []
    for idx, gene in enumerate(population):
        w, u, e = decode_gene(gene)
        loss = fitness(w, u, e)
        fitnesses.append((loss, gene))
        print(f"  Individual {idx}: w={w},u={u},e={e}  val_loss={loss:.4f}")
        if loss < best_loss:
            best_loss = loss
            best_gene = gene.copy()
            print(f"     New best!")

    # Selection: take top half
    fitnesses.sort(key=lambda x: x[0])
    parents = [g for (_, g) in fitnesses[:POP_SIZE//2]]

    # Reproduce: simple 2-point crossover + bitswap mutation
    new_pop = parents.copy()
    while len(new_pop) < POP_SIZE:
        p1, p2 = random.sample(parents, 2)
        # 2point crossover
        i1 = random.randint(1, GENE_LENGTH-2)
        i2 = random.randint(i1+1, GENE_LENGTH-1)
        child = p1[:i1] + p2[i1:i2] + p1[i2:]
        # mutation: flip a random bit
        if random.random() < 0.1:
            b = random.randrange(GENE_LENGTH)
            child[b] ^= 1
        new_pop.append(child)
    population = new_pop

#  5) REPORT BEST HYPERPARAMS & FINAL EVALUATION 
w_opt, u_opt, e_opt = decode_gene(best_gene)
print(f"\nBest hyperparams  window={w_opt}, units={u_opt}, epochs={e_opt}, loss={best_loss:.4f}")

# Retrain on Train+Val combined
X_tv = np.concatenate([X_train, X_val], axis=0)
y_tv = np.concatenate([y_train, y_val], axis=0)
X_tv_w, y_tv_w = sliding_windows(X_tv, y_tv, w_opt)
X_test_w, y_test_w = sliding_windows(X_test, y_test, w_opt)

final_model = build_model(w_opt, X_train.shape[2], u_opt, num_classes)
final_model.fit(X_tv_w, y_tv_w, epochs=e_opt, batch_size=32, verbose=1)

test_loss, test_acc = final_model.evaluate(X_test_w, y_test_w, verbose=0)
print(f"\nTest Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.2%}")
```
